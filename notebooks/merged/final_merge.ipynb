```python
import pandas as pd
```

```python
api_worldbank_dataset = pd.read_csv('api_worldbank_dataset.csv')
df_heritage_index_knoema = pd.read_csv('heritage_index_knoema.csv')
universities_data = pd.read_csv('universities_data.csv')
```

```python
merged_api = api_worldbank_dataset.merge(
    df_heritage_index_knoema,
    left_on='country_name',
    right_on='country',
    how='left'
)

merged_api['country_name'] = merged_api['country_name'].fillna(merged_api['country'])
merged_api = merged_api.drop(columns=['country'])

merged_api
```

```python
uni_df = universities_data.merge(
    merged_api,
    left_on='country',
    right_on='iso2_code',
    how='left'
)

uni_df
```

```python
uni_df.to_csv('universities_final_data.csv', index=False)
```
